```python
import requests
res = requests.get("https://api.airtable.com/v0/appNcYtL8fFZa1STA/iris?api_key=keyshdNC8CZdj1xgo")
print (res.text)
```

```python
import json
result = json.loads(res.text) #經過json.load後，從string轉為list type
result['records']
```

```python
import pandas as pd
df1 = pd.DataFrame(result['records'])
df1.head(3)
```

```python
df_id = pd.DataFrame(df1['id'].tolist())
df_id.columns = ['id']
df_id.head(3)
```

```python
df_ct = pd.DataFrame(df1['createdTime'].tolist())
df_ct.columns = ['createdTime']
df_ct.head(3)
```

```python
df_fields = pd.DataFrame(df1['fields'].tolist())
df_fields = df_fields[df_fields.columns[::-1]] #reverse columns order(option for view consistence)
df_fields.head(3)
```

```python
#欄位合併
df_iris = pd.merge( df_id, df_fields, left_index=True, right_index=True)
df_iris = pd.merge( df_iris, df_ct, left_index=True, right_index=True)
df_iris.head(3)
```
