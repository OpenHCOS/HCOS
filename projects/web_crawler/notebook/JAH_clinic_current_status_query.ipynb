```python
# http://www.jah.org.tw/netreg/opotime.asp?m=1&m1=4&m2=31&s=2&s1=12&s2=02&s3=0A09&s4=2

import pandas as pd
import requests
import re
from bs4 import BeautifulSoup

#說明:
# re.search:
# (.*) => .*  代表匹配除换行符之外的所有字符。
# (.*?)=> .*? 后面多个问号，代表非贪婪模式，也就是说只匹配符合条件的最少字符

sess=requests.Session()
url="http://www.jah.org.tw/netreg/opotime.asp"
params={
    "m":"1",
    "m1":"4",
    "m2":"31",
    "s":"2",     #院區
    "s1":"12",   #科別
    "s2":"02",   #診號
    "s3":"0A09", #(醫師ID)
    "s4":"2"     #(時段)
    }

html=sess.get(url, params=params)
soup=BeautifulSoup(html.content,'html.parser')
print(soup)
```

```python
item = soup.find_all("td", class_="T94E")
print(item.pop().get_text())
```
